```python
def read_file(file_name):
  with open('/content/drive/MyDrive/Corpus/{}'.format(file_name)) as f:
    lines=f.read()
  return lines
lines=read_file('Amazon.txt')
```

```python
def lowcase(inp):
  inp=inp.lower()
  return inp

def rem_punctuations(inp):
  inp = "".join([char for char in inp if char not in string.punctuation])
  inp = re.sub('[0-9]+', '', inp)
  return inp
text = rem_punctuations(lines)

def tokenization(inp):
  inp = inp.strip()
  inp = re.split('\W+', inp)
  return inp

text_tockens = tokenization(text)

def remove_singles(text):
  for i in text:
    if len(i)==1:
      text.remove(i)
  return text

text_tockens = remove_singles(text_tockens)

stopwords = nltk.corpus.stopwords.words('english')

def rem_Stopwords(inp):
  inp = [word for word in inp if word not in stopwords]
  return inp
text_tockens = rem_Stopwords(text_tockens)

def stemming(text):
  text = [nltk.PorterStemmer().stem(word) for word in text]
  return text

text_tockens = stemming(text_tockens)

def lemmatizer(text):
  text = [nltk.WordNetLemmatizer().lemmatize(word) for word in text]
  return text

text_tockens = lemmatizer(text_tockens)
```

```python
import glob
files = glob.glob("/content/drive/MyDrive/Corpus/*")
# extract all file names and creates a list
for x in range(len(files)):
  name = files[x].split("/")
  files[x] = name[-1] #get last element of list (only file name) print(files)
print(files)
```

```python

dic2 = {}
f=open("posting_list.txt","w")

for i in range(len(files)): 
  
  lines = read_file(files[i]) 
  lines = lowcase (lines) 
  lines = rem_punctuations(lines)
  tokens = tokenization(lines)
  tokens = remove_singles(tokens)
  tokens= rem_Stopwords(tokens)
  tokens = stemming(tokens) 
  tokens = lemmatizer(tokens)

  dic = {}
  
  # print(tokens)
  
  

  # below code will give term frequency for every single document seperately

  for word in tokens:
    if word in dic: 
      dic[word]+=1
    elif word not in dic.keys(): 
      dic[word] = 1
  # print(dic)

  
  # below code will give term frequency for every single word in one dictionary
  # for every document they appears in
  

  for word in dic.keys():
    if(word not in dic2.keys()):
      dic2[word] = {}
    dic2[word][i] = dic[word]
  
  
      
 
 
# below code will tell term, how many documents that term is in ,and how many times it appears in every document
#this will write that data in a seperate file
  
print(dic2)
for i in dic2:
  f.write(str(i)+", ["+str(len(dic2[i]))+str(dic2[i])+"]") 






```

```python
import math

N = len(files)
dic3 = {}
for i in dic2:  
  for x in dic2[i]:

    if dic2[i][x]==0:
      continue

    tf = 1 + math.log(dic2[i][x],10) 
    

    #creating a same dictionary as dic2 but changing tf-raw to tf-weight

    if i not in dic3.keys():
      dic3[i]={}                   

    # print(type(dic3[i][x]))

    dic3[i][x] = round(tf,2)

    # print(type(dic3[i][x]))
    

# in the code below we are normalizing the weights and re making the changes in dic3 only
for x in range(len(files)):
  sum = 0
  for i in dic3:
    # print(i,type(dic3[i][x]))
    if x in dic3[i]:
      sum += (dic3[i][x])**2
  for j in dic3:
    if x in dic3[j]: 
      dic3[j][x] = round(dic3[j][x]/(sum**0.5),2)
  


print(dic3)
```

```python



# query = input("tell me your query")
query = "Developing your Zomato business account and profile is a great way to boost your restaurants online reputation"
query = re.sub(r"\s+"," ",query) # substituting single or multiple spaces with single space 
    
lines = lowcase(query) 
lines = rem_punctuations (lines) 
tokens = tokenization (lines)
tokens = remove_singles (tokens)
tokens = rem_Stopwords (tokens)
tokens = stemming(tokens)
q_tokens = lemmatizer(tokens)
print(q_tokens)

wt = {}
qdf = {}


for i in range(len(files)): 
  
  lines = read_file(files[i]) 
  lines = lowcase (lines) 
  lines = rem_punctuations(lines)
  tokens = tokenization(lines)
  tokens = remove_singles(tokens)
  tokens= rem_Stopwords(tokens)
  tokens = stemming(tokens) 
  tokens = lemmatizer(tokens)

  dict_raw = {}
  
  
  # print(tokens)
  

  for word in q_tokens:
     

    #below code tells us that in how many documents/files is a term of our query is present 

    if word in tokens and word in qdf:
      qdf[word]+=1
    elif word in tokens and word not in qdf:
      qdf[word] = 1
    
    #below code tells us that in how many times is a term present in our query--------------term frequency  
    if word in dict_raw: 
      dict_raw[word]+=1
    elif word not in dict_raw.keys(): 
      dict_raw[word] = 1



print(dict_raw)
print(qdf)

#changing tf_raw into tf_weight and saving the changes in dict_raw only
for items in dict_raw:
  if dict_raw[items]!=0:
    dict_raw[items] = 1 + math.log(dict_raw[items],10)



#caculating idf

for items in qdf:
  if qdf[items]!=0:
    qdf[items] = round(math.log(N/qdf[items],10),2)



#calculating tf.idf
for word in q_tokens:
  wt[word] = round(dict_raw[word]*qdf[word] , 2)


print()
print("term-frequency..................", dict_raw)
print()
print("idf...........", qdf)
print()
print("tf-idf.............." , wt)


# below is the start of normalization of the tf-idf


sum = 0
for word in wt:
  sum += wt[word]**2
norm_tf_idf = {}

for word in wt:
  norm_tf_idf[word] = round(wt[word]/(sum**0.5),2)




print()
print("Normalized tf-idf..............." ,norm_tf_idf)





```

```python
import itertools

cosine = {}

#Calculating cosine similarity and ranking them 

for i in range(len(files)):
  for x in q_tokens:
    if i in dic3[x]:
      if document_no[i] in cosine:
        cosine[document_no[i]] += norm_tf_idf[x]*dic3[x][i]
      elif document_no[i] not in cosine:
        cosine[document_no[i]] = norm_tf_idf[x]*dic3[x][i]


print("cosine score.........", cosine)

#arranging the dictionary in decreasing cosine score

final_rank = dict(sorted(cosine.items(), key=lambda x:x[1], reverse=True))




print()
print()
ans_dict= dict(itertools.islice(final_rank.items(), 5))
print('\033[1;32m', ans_dict)



```
